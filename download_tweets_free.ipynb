Time to actually download the tweets! Make sure to specify the specific keywords (here, in the variable q). Using "OR" allows for the search of more than one keyword, using "AND" searches for keywords with both words.

For dates (since, until), times are always 12am of the specified date. For example setting since = "2019-07-22", until = "2019-07-23" will give tweets from July 22 at 12am to July 22 at 11:59pm.

Make sure to keep tweet_mode = "extended" to capture all the information associated with the given tweet.

Change the variable lang to change the language of the tweets.

Note that depending on the rate limit, you may have to wait up to 15 minutes for this step to finish running.

```python
# Get tweets using Tweepy api search
tweets = [status for status in tweepy.Cursor(api.search, 
                                             q= "$GLD OR #GLD OR $GOLD", #also try: $GLD OR #GLD OR $GOLD, $GDX OR GDX, $JNUG or JNUG, #GDXJ OR GDXJ 
                                             since = "2019-07-25",
                                             until = "2019-07-25",
                                             tweet_mode="extended",
                                             lang="en",
                                            ).items()]
```

Now that we have the tweets, we want to save them into a json file! This will allow us to retain as much information as possible.

```python
import json
with open('2019-07-25.json', 'a') as f:
    for tweet in tweets:
        json.dump(tweet._json,f)  
        f.write('\n')
```

Let's reload the json file so we can format the information and save it as an easy to work with csv file.

```python
tweet_files = '2019-07-25.json'
tweets = []

with open(tweet_files, 'r') as f:
    for line in f.readlines():
        tweets.append(json.loads(line))
```

We're going to format the data using a dataframe with the columns of information that we want to see. Here, I use id, name, date, text, and location. Check out the Twitter API documentation for more information you can store.

```python
import pandas as pd
df = pd.DataFrame()
df['id'] = list(map(lambda tweet: tweet['id'], tweets))
df['name'] = list(map(lambda tweet: tweet['user']['screen_name'], tweets))
df['date'] = list(map(lambda tweet: tweet['created_at'], tweets))
tweet_text = []

for p in tweets:
    if 'retweeted_status' in p:
        string = p['retweeted_status']['full_text']
        string.encode('utf-8', 'replace')
        tweet_text.append(string)           
    else:
        string = p['full_text']
        string.encode('utf-8', 'replace')
        tweet_text.append(p['full_text'])
        
df['text'] = tweet_text
#f['text'] = list(map(lambda tweet: tweet['full_text'], tweets))
df['location'] = list(map(lambda tweet: tweet['user']['location'], tweets))
```
